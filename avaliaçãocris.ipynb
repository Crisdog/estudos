2. Desafio de criptografia. De uma forma simples, a criptografia é uma forma de
esconder uma mensagem substituindo seus caracteres originais por outros de
modo que a mensagem fique incompreensível para quem não conhece a cifra.
Existem muitas formas de se encriptar uma mensagem. Uma muito antiga e
conhecida é a cifra de César. A cifra de César foi um dos primeiros algoritmos de
criptografia existentes. Esse nome foi atribuído em homenagem a Júlio César, que a
utilizava quando queria enviar mensagens de cunho militar de uma forma que a real
informação permanecesse escondida mesmo se a mensagem fosse interceptada.
Esse padrão de encriptação consiste na substituição de um caractere por outro à
sua frente ou anterior. Por exemplo, se usarmos o mapeamento de 3 posições, todas
as vezes que a letra ‘a’ aparecer devemos substituí-la por ‘d’. Para nosso desafio
vamos implementar a cifra de César com um mapeamento de 5 posições. Quer dizer
que toda vez que a palavra contiver uma letra ‘a’ na palavra cifrada deverá aparecer
‘f’. Informações essenciais e dados gerais:
Array base do alfabeto: cifra = ['a', 'b', 'c', 'd', 'e', 'f', 'g',
'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's',
't', 'u', 'v', 'w', 'x', 'y', 'z', '1', '2', '3', '4', '5']
Você deve desenvolver uma função que recebe uma palavra e imprime essa palavra
criptografada na tela.

```python
def cifra_de_cesar (palavra):
  cifra = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','m', 'n', 'o', 'p', 'q', 'r', 's',
't', 'u', 'v', 'w', 'x', 'y', 'z', '1', '2', '3', '4', '5']
  palavra_cifrada = ""

  for letra in palavra:
    if letra.lower()in cifra:
      indice_letra = cifra.index (letra.lower())
      indice_letra_cifrada = (indice_letra + 5)% len(cifra)
      letra_cifrada = cifra[indice_letra_cifrada]
      palavra_cifrada += letra_cifrada
    else:
        palavra_cifrada += letra_cifrada

  print ("palavra cifrada", palavra_cifrada)  

p =input("digite uma palavra")

cifra_de_cesar(p)
 
```
